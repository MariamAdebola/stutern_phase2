```python
#import libraries

import pandas as pd
from sqlalchemy import create_engine
import os
import shutil
```

```python
#create a database connection
engine = create_engine("postgresql+psycopg2://postgres:4991@localhost:5432/udemy_course_database", pool_recycle=-1)
print("connection successful")
```

```python
#create a connection to run sql queries
db_engine = engine.connect()
print("connection successful")
```

```python
#create a function to run queries
def db_query(query:str, db_conn, index_col='course_id'):
    df = pd.read_sql(query, db_conn, index_col)
    return df
```

**Question 1: Write a query for paid courses, showing the  course with the highest number of subscribers per month**

```python
query = """SELECT month, subject, MAX(total_subscribers) AS highest_subscribers
    FROM 
    (
        SELECT DISTINCT subject, SUM(num_subscribers:: INTEGER) AS total_subscribers,
        to_char(published_timestamp, 'Month') AS month
        FROM udemy_course_table
        WHERE is_paid='TRUE'
        GROUP BY subject, month
    ) AS sub_table
GROUP BY 1, 2
ORDER BY 1"""

df = db_query(query, db_engine, index_col='subject')
df
```

**Question 2: Write a query for paid courses, showing the course with the highest sales per month**"

```python
query = """SELECT month, subject, total_sales, MAX(total_subscribers) AS highest_subscribers
    FROM 
    (
        SELECT DISTINCT subject, SUM(num_subscribers:: INTEGER) AS total_subscribers, SUM(price) AS total_sales,
        to_char(published_timestamp, 'Month') AS month
        FROM udemy_course_table
        WHERE is_paid='TRUE'
        GROUP BY subject, month
    ) AS sub_table
GROUP BY 1, 2, 3
ORDER BY 1"""

df = db_query(query, db_engine, index_col='subject')
df
```
